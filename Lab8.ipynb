## Resizing Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('datasets/cover1.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

nearest_inter = cv2.resize(image, (250,250),interpolation = cv2.INTER_NEAREST)
linear_inter = cv2.resize(image, (250,250),interpolation = cv2.INTER_LINEAR)
area_inter = cv2.resize(image, (250,250),interpolation = cv2.INTER_AREA)
cubic_inter = cv2.resize(image, (250,250),interpolation = cv2.INTER_CUBIC)
lanczos_inter = cv2.resize(image, (250,250),interpolation = cv2.INTER_LANCZOS4)

names = ['Original Image','INTER_NEAREST', 'INTER_LINEAR','INTER_AREA', 'INTER_CUBIC', 'INTER_LANCZOS4']
images = image, nearest_inter, linear_inter, area_inter, cubic_inter, lanczos_inter

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i])
    plt.title(names[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

## Rotating Images

```python
image = cv2.imread('datasets/cover2.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
rows,cols = image.shape[:2] 
print(image.shape)

M = cv2.getRotationMatrix2D((cols/2,rows/2),20,1) 
dst = cv2.warpAffine(image,M,(cols,rows))

plt.subplot(1,2,1),plt.imshow(image)
plt.title('Original Image')
plt.xticks([]),plt.yticks([])

plt.subplot(1,2,2),plt.imshow(dst)
plt.title('Rotated Image')
plt.xticks([]),plt.yticks([])

# plt.imshow(dst)
# plt.show()
# cv2.imwrite('datasets/cover7.jpg',dst)
```

```python
image = cv2.imread('datasets/cover4.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
rows,cols = image.shape[:2]
#(col/2,rows/2) is the center of rotation for the image
# M is the cordinates of the center
M = cv2.getRotationMatrix2D((cols/2,rows/2),180,1)
dst = cv2.warpAffine(image,M,(cols,rows))
# plt.imshow(dst)
# plt.show()

plt.subplot(1,2,1),plt.imshow(image)
plt.title('Original Image')
plt.xticks([]),plt.yticks([])

plt.subplot(1,2,2),plt.imshow(dst)
plt.title('Rotated Image')
plt.xticks([]),plt.yticks([])


# cv2.imwrite('datasets/cover9.jpg',dst)
```
